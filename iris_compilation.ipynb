# 6. MODELLING

## K Means - Clustering

### 6.1 Construct the model

This part depends a lot on the model to apply, in this case we're going to begin with KMEANS

```python
## ELBOW METHOD

from sklearn.cluster import KMeans

inertias = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", n_init=10, random_state=42)
    #kmeans.fit(X=X_train)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Visualizing the ELBOW method to get the optimal value of K
plt.plot(range(1, 11), inertias)
plt.title("The Elbow Method")
plt.xlabel("# of clusters")
plt.ylabel("inertia")
plt.show()
```

We can say that K = 3 or 4, we can take 3 

```python
k = 3

Model_kmeans = KMeans(n_clusters=k, init="k-means++", n_init=10, random_state=42)
Model_kmeans.fit(X)
```

### 6.2 Evaluate the model

```python
clusters = Model_kmeans.predict(X)
print("labels: ", Model_kmeans.labels_)
print("Clusters: ", clusters)
print("Target: ", irisTarget)
```

```python
# Impact of C

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_classification

# Crear datos con ruido
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0,
                           n_informative=2, n_clusters_per_class=1,
                           flip_y=0.1, random_state=42)

# Diferentes valores de C
C_values = [0.1, 1, 10, 100]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for idx, C_val in enumerate(C_values):
    # Entrenar SVM
    clf = svm.SVC(kernel='rbf', C=C_val, gamma='scale')
    clf.fit(X, y)
    
    # Crear mesh para visualización
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot
    ax = axes[idx]
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', 
               edgecolors='k', s=50)
    
    # Marcar support vectors
    ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
               s=200, linewidth=2, facecolors='none', edgecolors='red',
               label='Support Vectors')
    
    score = clf.score(X, y)
    n_support = len(clf.support_vectors_)
    
    ax.set_title(f'C = {C_val}\nAccuracy: {score:.2f}\n'
                f'Support Vectors: {n_support}', fontsize=12)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Observa:")
print("C pequeño → Frontera suave, más support vectors")
print("C grande → Frontera compleja, menos support vectors")
```

```python
# Importance of gamma

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_circles

# Datos en círculos
X, y = make_circles(n_samples=100, noise=0.1, factor=0.5, random_state=42)

# Diferentes valores de gamma
gamma_values = [0.1, 1, 10, 100]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for idx, gamma_val in enumerate(gamma_values):
    # Entrenar SVM
    clf = svm.SVC(kernel='rbf', C=1.0, gamma=gamma_val)
    clf.fit(X, y)
    
    # Crear mesh
    h = 0.02
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot
    ax = axes[idx]
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', 
               edgecolors='k', s=50)
    
    score = clf.score(X, y)
    n_support = len(clf.support_vectors_)
    
    ax.set_title(f'gamma = {gamma_val}\nAccuracy: {score:.2f}\n'
                f'Support Vectors: {n_support}', fontsize=12)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Observa:")
print("gamma bajo → Frontera suave y simple")
print("gamma alto → Frontera compleja, se ajusta a cada punto")
```
